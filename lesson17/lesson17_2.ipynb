```python
import numpy as np
import pandas as pd
scores = np.random.randint(50,101,size=(50,5)) #50分到~100分 size=50,5 50個列 5個欄
scores_df = pd.DataFrame(scores,index=range(1,51),columns=['國文', '英文', '數學', '地理', '歷史'])
scores_df.index.name = "學號"
scores_df.columns.name = "科目"
scores_df
```

```python
scores_df1 = scores_df.copy()
```

```python
def cal1(s):        #s是Series  直行的是Series ex: 國文1~50
    return s.sum()
scores_df.apply(cal1,axis=1)    #axis=1 是橫的 ex: 1號 國文75 英文81 數學89	地理92 歷史85  
```

```python
s1 = scores_df.iloc[1]    #iloc[0] 抓第一筆
int(s1[s1<70].count())         #count() 統計數量
```

```python
def cal2(s):
    nums = int(s[s<60].count())        
    return nums
scores_df1['不及格的科目'] = scores_df.apply(cal2,axis=1) 
scores_df1
```

```python
#5個Series傳出成DataFrame
def cal3(s):
    return pd.Series([s.min(),s.max(),s.std()],index=['最低分','最高分','標準差'])
scores_df3 = scores_df.apply(cal3)          #apply(function)
```
