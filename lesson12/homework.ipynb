```python
import random
import matplotlib.pyplot as plt

#製作figure  
figure = plt.figure(figsize=(10,5)) #figsize設定寬跟高(x,y)
#在figure上製作一個axes
axes = figure.add_subplot(1, 1, 1)

days = [num for num in range(1,31)] 
celsius = [round(random.uniform(28.5,35.2),1) for _ in range(1,31)] 

axes.grid(axis='both')  #產生網格
axes.set_title('july weather') #設定標題
axes.set_xlabel('Days') #設定x軸名稱
axes.set_ylabel('Celsius') #設定y軸名稱
axes.set_xticks(days) #設定x軸刻度 顯示1~30
axes.plot(days, celsius,'ro--')
#顯示圖
plt.show()
```
