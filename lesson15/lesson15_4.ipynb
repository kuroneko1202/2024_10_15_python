```python
import pandas as pd
df = pd.read_csv('world.csv')
df1 = df[['洲名','國家','日期','總確診數','新增確診數','總人口數','解封指數']]
country = input("請輸入國家")
country_df = df1.query('國家==@country') 
country_max = int(country_df['新增確診數'].max())
print(f"{country}的最大新增確診數為：{country_max}")
country_df.query('新增確診數 == @max')
```
