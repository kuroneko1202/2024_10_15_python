```python
import matplotlib.pyplot as plt

# 1.製作figure  
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y,'ro-')

# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
axes = fig.add_subplot(1, 1, 1)
subjects = ['chinese','english','math','history','discover']
nums1 = [1, 4, 9, 16, 25]
nums2 = [25, 16, 12, 4, 1]
axes.plot(subjects,nums1,'+--y',label='student1')
axes.plot(subjects,nums2,'o-r',label='student2')
axes.set_title('students')
axes.set_xlabel('subject')
axes.set_ylabel('scores')
axes.grid(axis='y')
axes.set_xticks([0, 1, 2, 3, 4])
axes.set_xticklabels(['ch','en','ma','hi','cover'])
axes.legend()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# 1.製作figure  
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)

days = [num for num in range(1,31)]
celsius = [round(random.uniform(28.5,35.2),1) for _ in range(1,31)]

axes.grid(axis='both') 
axes.set_title('july weather')
axes.set_xlabel('Days')
axes.set_ylabel('Celsius')
axes.set_xticks(days)
axes.plot(days, celsius,'ro--')
# 4.顯示圖
plt.show()
```
